# Modeling an Infection with Changing R for San Diego County

The SIR infection model is a simple way of exploring the effects of a social distance policy on a large population such as San Diego County

The SIR infection model is a simple set of differential equations that describe how a disease can spread under very homogeneous, stable conditions. The model gets it's name from the three "compartments" for people in diferent conditions: 

* Susceptible: people who can be infected.
* Infected: people who are infected.
* Recovered or Removed: people who were infected, and can't get infected again. 

There are [many good overviews](https://www.maa.org/press/periodicals/loci/joma/the-sir-model-for-spread-of-disease-the-differential-equation-model) of the model online, so in this brief analysis, we will just use the model to visuaize what happens when the $R$ parameter for an epidemic changes. 

$R$ is the ["effective reproduction number"](https://www.healthknowledge.org.uk/public-health-textbook/research-methods/1a-epidemiology/epidemic-theory) of a disease, which describes the rate of increase of infection for curent conditions. It differs from the similar number, one most people are more familliar with from recent press, the basic reproduction number $R_0$, in that $R$ can change, but $R_0$ cannot. 

The SIR model doesn't use $R$ or $R_0$ directly; it uses an infection rate $\beta$ and recovery rate $\gamma$, and $R_0$ = $\frac{\beta N}{\gamma}$, where $N$ is the total population.  In this simulation, we'll use the SIR model to see what happens when a popuation the size of San Diego County gets one infection, with one case with constant $R_0$ and other cases where $R$ changes from $R_0$ to a lower value. We will change $R$ by changing $\beta$ and leaving $\gamma$ constant, which is a simple simulation of San Diego impementing a social distancing policy.

```python
from functools import partial 

#
# This is the main function for calculating a set in the solution. 
def sir_diff( N, beta, gamma, sir, t):
    """Calculate new difference values for input values
    
        sir[0] == S
        sir[1] == I
        sir[2] == R
    
    """
  
    dsdt = - (beta * sir[0] * sir[1])/N
    didt = (beta * sir[0] * sir[1])/N - gamma * sir[1]
    drdt = gamma * sir[1]
    
    return [dsdt, didt, drdt]


def make_sir_df(N, t, sir, kn_rate=.15):
    """Build a nice dataframe from the Numpy array"""
    df = pd.DataFrame({
        't' : t,
        'S' : sir[:, 0],
        'I' : sir[:, 1],
        'R' : sir[:, 2],
    }).set_index('t')

    # The simulation produces man points per day, but it is more convenient
    # to have just one row per day. 
    df = df.groupby(np.round(df.index).astype(int)).mean()

    #df['known_cases'] = (N-df.S)*kn_rate
    # San Diegos data starts with 8 cases, so make that day zero
    #day_0_idx = (df.known_cases-8).abs().idxmin()
    #df = df.shift(-day_0_idx)

    return df

def solve_sir_o(days, N, beta, gamma, points=1000):
    """Solve it with ODEINT"""
    #N = 3.2e6
    S = N - 1
    I = 1 # Must be at least one infection
    R = 0
 
    sir0 = (S, I, R) # initial conditions

    t = np.linspace(0, days, num=points)

    # solve ODE
    # the parameters are, the equations, initial conditions, 
    # and time steps (between 0 and 100)
    
    # This partial will be a new function that calls
    # sir_diff with fixed values for beta and gamma
    f = partial(sir_diff, N, beta, gamma)
    
    return t, odeint(f, sir0, t)

    
def run_sir_o(N, beta, gamma, days, points):
    """Run the simulation and solve with odeint"""
    t, s = solve_sir_o(days, N, beta, gamma, points)
    
    return make_sir_df(N, t, s)
    

```

```python
fig, ax = plt.subplots(figsize=(15,12))
fig.suptitle("SIR Model for San Diego County\nwith Fixed R0 and Reducing R0", fontsize=20, y=1.05);

N = 3.2e6 
gamma =  .3

r1 = 3
r2 = .9
beta1 = r1 * gamma
beta2 = r2 * gamma

days = 45
points = 1000

# Fixed

df_i = run_sir_o(N, beta1,  gamma, days, points)
ax.set_title(f'R0 = {r1}. Max Infections = {df_i.I.max().round(0).astype(int)}', fontsize=14)
df_i["R"] = df_i["R"]/2000
df_i.plot(ax=ax, linewidth=4)
ax.set_ylabel('Number of People')
ax.set_xlabel('Days since first infection')
#ax.set_xlim(15,days)
ax.set_ylim(0,2000)
```

```python
def solve_sir_i(t, N, beta, gamma):
    """Solve the simulation by iterating over the time slices"""
    params = (N-1, 1, 0)
    
    dv = len(t)/max(t)
    
    for i, t_ in enumerate(t):
              
        try:
            b = beta[i]
        except TypeError:
            b = beta
        
        try:
            g = gamma[i]
        except TypeError:
            g = gamma
        
        
        diff = np.array(sir_diff(N, b, g, params, t_))/dv
        yield params
        params = np.add(params, diff)
 
def run_sir_i(N, beta, gamma, days, points):
    """Solve it by iterating. Not as accurate, but allows beta and gama to change. """

    
    t = np.linspace(0, days, num=points)
    s = np.array(list(solve_sir_i(t, N, beta, gamma)))
    df_i = make_sir_df( N, t, s)
    return df_i

def make_beta(r1, t1, r2, t2, pdf, points):
    beta = np.concatenate( (np.array([beta1]*int(t1*pdf)),  
                        np.linspace(beta1, beta2, num=int(t1+t2)*pdf),  
                        np.array([beta2]*int(points-((t1+t2)*pdf)))), axis=0 ) 
    return beta
```
